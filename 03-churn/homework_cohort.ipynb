```python
import pandas as pd
import numpy as np
```

```python
# !wget https://raw.githubusercontent.com/alexeygrigorev/mlbookcamp-code/master/chapter-02-car-price/data.csv
```

```python
df = pd.read_csv('data.csv')
df.T
```

```python
used_columns = ['Make',
'Model',
'Year',
'Engine HP',
'Engine Cylinders',
'Transmission Type',
'Vehicle Style',
'highway MPG',
'city mpg',
'MSRP']

df = df[used_columns]
df.T
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.T
```

```python
df.isnull().sum()
```

```python
df.engine_hp = df.engine_hp.fillna(0)
df.engine_cylinders = df.engine_cylinders.fillna(0)
```

```python
df.isnull().sum()
```

```python
df['price'] = df.msrp
```

```python
del df['msrp']
df.T
```

```python
df.transmission_type.mode()
```

```python
numerical_cols = []
numerical_cols.append(list(df.dtypes[df.dtypes == 'int64'].index))
numerical_cols.append(list(df.dtypes[df.dtypes == 'float64'].index))

numerical_cols = numerical_cols[0] + numerical_cols[1]
numerical_cols
```

```python
df[numerical_cols].corr().unstack().sort_values(ascending=False)
```

```python
mean_price = df.price.mean()
df['above_average'] = np.where(df.price > mean_price, True, False)
# del df['price']
df.T
```

```python
#setup the framework 🥳
from sklearn.model_selection import train_test_split
```

```python
df_train_full, df_test = train_test_split(df, test_size=0.20, random_state=42)
df_train, df_val = train_test_split(df_train_full, test_size=0.25, random_state=42)

len(df_train_full), len(df_train), len(df_test), len(df_val)
```

```python
y_train_full = df_train_full.above_average.values
y_train = df_train.above_average.values
y_val = df_val.above_average.values
y_test  = df_test.above_average.values
```

```python
from sklearn.metrics import mutual_info_score
```

```python
df_train.T
```

```python
def mutal_info(series):
    return mutual_info_score(df_train.above_average, series)
```

```python
mi = df_train.apply(mutal_info)
```

```python
mi.sort_values(ascending=False).round(2)
```

```python
# for safety delete the Target variable.
del df_train_full['above_average']
del df_train['above_average']
del df_val['above_average']
del df_test ['above_average']

del df_train_full['price']
del df_train['price']
del df_val['price']
del df_test ['price']
```

```python
# WE are are going to do one hot encoding for this dataset
from sklearn.feature_extraction import DictVectorizer

train_dict = df_train.to_dict(orient='records')
dict_vector = DictVectorizer(sparse=False)
dict_vector.fit(train_dict)
X_train = dict_vector.transform(train_dict)

X_train
```

```python
## Training the model 

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='liblinear', C=10, max_iter=1000, random_state=42)
model.fit(X_train, y_train)
```

```python
## Now check it with validation data 

val_dict = df_val.to_dict(orient='records')
X_val = dict_vector.transform(val_dict)

y_pred = model.predict(X_val)

## Checking the accuracy
from sklearn.metrics import accuracy_score

accuracy = np.round(accuracy_score(y_pred, y_val),2)
accuracy
```

```python
features = list(df_train.dtypes.index.values)
features
```

```python
orig_score = accuracy

for c in features: 
    subset = features.copy()
    subset.remove(c)
    
    train_dict = df_train[subset].to_dict(orient='records')
    
    dv = DictVectorizer(sparse=False)
    dv.fit(train_dict)
    
    X_train = dv.transform(train_dict)
    
    model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)
    model.fit(X_train, y_train)
    
    val_dict = df_val[subset].to_dict(orient='records')
    X_val = dv.transform(val_dict)
    
    y_pred = model.predict(X_val)
    
    score = accuracy_score(y_val, y_pred)
    print(c, orig_score - score, score)
```

```python
#@ SPLITTING THE DATASET:
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df_train_full, test_size=0.25, random_state=42)
```

```python
#@ PREPARING THE DATASET:
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
y_train_full = df_train_full.above_average.values
y_train = df_train.above_average.values
y_val = df_val.above_average.values
y_test  = df_test.above_average.values
```

```python
# for safety delete the Target variable.
del df_train_full['above_average']
del df_train['above_average']
del df_val['above_average']
del df_test ['above_average']

del df_train_full['price']
del df_train['price']
del df_val['price']
del df_test ['price']
```

```python
train_dict = df_train.to_dict(orient='records')
```

```python

#@ VECTORIZING THE DATASET:
dv = DictVectorizer(sparse=False)
dv.fit(train_dict)

X_train = dv.transform(train_dict)

val_dict = df_val.to_dict(orient='records')
X_val = dv.transform(val_dict)
     
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error

for a in [0, 0.01, 0.1, 1, 10]:
    model = Ridge(alpha=a, solver="sag", random_state=42)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_val)
    
    score = np.sqrt(mean_squared_error(y_val, y_pred))
    
    print(a, round(score, ))
```
